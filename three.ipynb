```kotlin
import java.io.File

val input = File("three.txt").readLines()
```

```kotlin
tailrec fun getLargest(left: Int, right: Int, input: String): Int {
    if (input.isEmpty()) return left*10 + right
    val new = input.get(0).digitToInt()
    if (right > left) return getLargest(right, new, input.drop(1))
    if (new > right) return getLargest(left, new, input.drop(1))
    return getLargest(left, right, input.drop(1))
}

val part1 = input.sumOf { bank ->
    getLargest(bank.get(0).digitToInt(), bank.get(1).digitToInt(), bank.drop(2))
}

tailrec fun getLargest(input: String, result: String, size: Int): Long {
    if (size == 0) return result.toLong()

    val search = input.dropLast(size - 1)
    val largestPosition = search.indexOf(search.max())
    return getLargest(input.drop(largestPosition + 1), result + input[largestPosition], size - 1)
}

val part2 = input.sumOf { bank ->
    getLargest(bank, "", 12)
}




```
