```kotlin
import java.io.File

typealias Shape = List<List<Char>>  // A 2D grid representing a present shape
typealias ShapeCounts = List<Int>    // List of quantities for each shape index
typealias Dimensions = Pair<Int, Int> // Width and height of a region (width, height)
typealias Region = Pair<Dimensions, ShapeCounts>  // dimensions + shape counts

val input = File("12.txt").readText().split("\r\n\r\n").let { blocks ->
    Pair(
        blocks.dropLast(1).map { block ->
            block.split("\r\n").drop(1).map { line ->
                line.toCharArray().toList()
            }
        },
        blocks.takeLast(1)[0].split("\r\n").dropLast(1).map { line ->
            val (dim, nums) = line.split(": ")
            val (width, height) = dim.split("x").map { it.toInt() }
            Region(
                 Pair(width, height),
                 nums.split(" ").map { it.toInt() }
            )
        }
    )
}
```

```kotlin
typealias Shape = List<List<Char>>  // A 2D grid representing a present shape
typealias ShapeCounts = List<Int>    // List of quantities for each shape index
typealias Dimensions = Pair<Int, Int> // Width and height of a region (width, height)
typealias Region = Pair<Dimensions, ShapeCounts>  // dimensions + shape counts

// taken by https://github.com/scully-dana/advent-of-code/blob/master/2025/src/Day12.kt
fun lazyFit(regionDimensions: Dimensions, shapeCounts: ShapeCounts): Boolean {
    val (width, height) = regionDimensions
    val area = (width / 3) * (height / 3)
    val totalPresents = shapeCounts.sum()
    return (totalPresents <= area)
}

fun part1(regions: List<Region>): Int {
    return regions.count { region: Region ->
        lazyFit(region.first, region.second)
    }
}

val part1 = part1(input.second)
```
