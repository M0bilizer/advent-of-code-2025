```kotlin
import java.io.File

val input = File("2.txt").readLines().first().split(",")
```

```kotlin
fun process(s: String): Pair<Long, Long> {
    val range = s.split("-")
    return Pair(range[0].toLong(),range[1].toLong())
}

tailrec fun isEqual(left: String, right: String): Boolean {
    if (left.length != right.length) return false;
    if (left.length == 0 && right.length == 0) return true;
    if (left.get(0) != right.get(0)) return false
    return isEqual(left.drop(1), right.drop(1))
}

val part1 = input.sumOf { it ->
    val (low, high) = process(it)
    (low..high).sumOf { id ->
        val idStr = id.toString()
        val mid = idStr.length / 2
        if (isEqual(idStr.substring(0, mid), idStr.substring(mid))) {
            id.toLong()
        } else {
            0
        }
    }
}

fun isRepeatedPattern(str: String): Boolean {
    return str.length > 1 && (str + str).indexOf(str, 1) < str.length
}

val part2 = input.sumOf { it ->
    val (low, high) = process(it)
    (low..high).sumOf { id ->
        val idStr = id.toString()
        if (isRepeatedPattern(idStr)) {
            id.toLong()
        } else {
            0
        }
    }
}
```
