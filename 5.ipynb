```kotlin
import java.io.File

val input = File("5.txt").readLines()
    .let { lines ->
        val splitIndex = lines.indexOf("")
        Pair(
            lines.take(splitIndex).map { line ->
                line.split("-").map { it.toLong() }.let { (start, end) -> start..end }
            },
            lines.drop(splitIndex + 1).map { it.toLong() }
        )
    }
```

```kotlin
fun part1(ranges: List<LongRange>, ids: List<Long>): Int {
    var count = 0
    for (id in ids) {
        if (ranges.any { it -> id in it})
            count++
    }
    return count;
}

val part1 = part1(input.first, input.second)

fun part2(ranges: List<LongRange>): Long {
    val sorted = ranges.sortedBy { it.first }
    val merged = mutableListOf<LongRange>()

    sorted.forEach { current ->
        if (merged.isEmpty() || current.first > merged.last().last + 1) {
            merged.add(current)
        } else {
            val last = merged.removeAt(merged.lastIndex)
            merged.add(last.first..maxOf(last.last, current.last))
        }
    }
    return merged.sumOf { it.last - it.first + 1 }
}

val part2 = part2(input.first)
```
