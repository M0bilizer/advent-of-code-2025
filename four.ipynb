```kotlin
import java.io.File

val input = File("four.txt").readLines().map {it.toList()}
```

```kotlin
fun getAdjacencyMatrix(input: List<List<Char>>): List<List<Int>> {
    val adjacencyMatrix = List(input.size) { i ->
        MutableList(input[0].size) { j ->
            if (input[i][j] == '.') -1 else 0
        }
    }
    val directions = listOf(
        -1 to -1, -1 to 0, -1 to 1,
        0 to -1,          0 to 1,
        1 to -1,  1 to 0, 1 to 1
    )
    for (i in input.indices) {
        for (j in input[i].indices) {
            if (input[i][j] == '@') {
                for ((di, dj) in directions) {
                    val ni = i + di
                    val nj = j + dj
                    if (ni in input.indices && nj in input[i].indices) {
                        if (adjacencyMatrix[ni][nj] != -1) {
                            adjacencyMatrix[ni][nj]++
                        }
                    }
                }
            }
        }
    }
    return adjacencyMatrix
}

val part1 = getAdjacencyMatrix(input).flatten().count { it < 4 && it != -1 }

tailrec fun recursivelyRemoveRoll(input: List<List<Char>>, totalRemoved: Int): Int {
    val matrix = getAdjacencyMatrix(input)
    val currentRemovals  = matrix.flatten().count { it < 4 && it != -1 }
    if (currentRemovals  == 0) return totalRemoved
    else {
        val nextInput = input.mapIndexed { i, row ->
            row.mapIndexed { j, char ->
                if (matrix[i][j] < 4) '.' else char
            }
        }
        return recursivelyRemoveRoll(nextInput, totalRemoved + currentRemovals)
    }
}

val part2 = recursivelyRemoveRoll(input, 0)
```
