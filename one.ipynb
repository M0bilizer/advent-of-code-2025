```kotlin
import java.io.File

val input = File("one.txt").readLines()
```

```kotlin
class Counter(var state: Int) {
    fun update(step: Int) {
        this.state += step
        if (state > 99) {
            this.state = state % 100
        }
        if (state < 0) {
            this.state = 100 + state
        }
    }
}

fun process(s: String): Int {
    val (direction, step) = s.take(1) to s.drop(1).toInt()
    if (direction.startsWith("R")) {
        return step % 100
    } else {
        return -(step % 100)
    }
}

val c = Counter(50)
var zero = 0
input.forEach { it -> c.update(process(it)); if (c.state == 0) zero++}
zero
```
