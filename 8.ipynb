```kotlin
import java.io.File

val input = File("8.txt").readLines()
    .map { line ->
        val parts = line.split(",").map { it.trim().toInt() }
        Triple(parts[0], parts[1], parts[2])
    }
```

```kotlin
fun getSortedDistances(coordinates: List<Triple<Int, Int, Int>>): List<Pair<Int, Int>> {
    val distances = mutableListOf<Pair<Pair<Int, Int>, Long>>()

    for (i in coordinates.indices) {
        for (j in i + 1 until coordinates.size) {
            val (x1, y1, z1) = coordinates[i]
            val (x2, y2, z2) = coordinates[j]

            val dx = x1 - x2
            val dy = y1 - y2
            val dz = z1 - z2
            val distSquared = dx.toLong() * dx + dy.toLong() * dy + dz.toLong() * dz
            distances.add((i to j) to distSquared)
        }
    }

    return distances.sortedBy { it.second }
        .map { it.first }
}

fun part1(coordinates: List<Triple<Int, Int, Int>>, distances: List<Pair<Int, Int>>): Int {
    val groups = coordinates.indices.map { setOf(it) }.toMutableList()
    distances.take(1000).forEachIndexed { index, (i, j) ->
        val groupI = groups.indexOfFirst { it.contains(i) }
        val groupJ = groups.indexOfFirst { it.contains(j) }
        if (groupI != groupJ) {
            val merged = groups[groupI] union groups[groupJ]
            val (first, second) = if (groupI > groupJ) groupI to groupJ else groupJ to groupI
            groups.removeAt(first)
            groups.removeAt(second)
            groups.add(merged)
        }
    }
    return groups.sortedByDescending { it.size }.take(3).map {it -> it.size}.reduce { acc, ele -> acc * ele }
}

fun part2(coordinates: List<Triple<Int, Int, Int>>, distances: List<Pair<Int, Int>>): Long {
    val groups = coordinates.indices.map { setOf(it) }.toMutableList()
    var last: Pair<Int, Int> = distances[0]
    distances.forEachIndexed { index, (i, j) ->
        val groupI = groups.indexOfFirst { it.contains(i) }
        val groupJ = groups.indexOfFirst { it.contains(j) }
        if (groupI != groupJ) {
            last = Pair(i,j)
            val merged = groups[groupI] union groups[groupJ]
            val (first, second) = if (groupI > groupJ) groupI to groupJ else groupJ to groupI
            groups.removeAt(first)
            groups.removeAt(second)
            groups.add(merged)
        }
    }
    return coordinates[last.first].first.toLong() * coordinates[last.second].first.toLong()
}

val part1 = part1(input, getSortedDistances(input))
val part2 = part2(input, getSortedDistances(input))
```
