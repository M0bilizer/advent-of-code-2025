```kotlin
fun process(s: String): Int {
    val (direction, step) = s.take(1) to s.drop(1).toInt()
    if (direction.startsWith("R")) {
        return step
    } else {
        return -(step)
    }
}

tailrec fun part1(count: Int, dial: Int, input: List<String>): Int {
    if (input.isEmpty()) return count
    val new = (dial + process(input[0])).mod(100)
    if (new == 0)
        return part1(count + 1, 0, input.drop(1))
    else
        return part1(count, new, input.drop(1))
}

tailrec fun part2(count: Int, dial: Int, input: List<String>): Int {
    if (input.isEmpty()) return count
    val moveBy = process(input[0])
    val passThroughZero = if (moveBy > 0) {
        (dial + moveBy) / 100
    } else {
        val endPos = (dial + moveBy).mod(100)
        val wrapCount = when {
            dial == 0 -> 0
            endPos == 0 -> 1
            endPos > dial -> 1
            else -> 0
        }
        (-moveBy / 100) + wrapCount
    }
    val new = (dial + process(input[0])).mod(100)
    return part2(count + passThroughZero, new, input.drop(1))
}

val part1 = part1(0, 50, input)
val part2 = part2(0, 50, input)
```
